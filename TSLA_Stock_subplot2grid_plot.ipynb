```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
#import pandas_datareader.data as web

style.use('ggplot')
```

```python
#start = dt.datetime(2019, 1, 1)
#end = dt.datetime.now()
#df = web.DataReader("TSLA", 'morningstar', start, end)
data = pd.read_csv("TSLA.csv", parse_dates=True, index_col=0)
#data.reset_index(inplace=True)
#data.set_index("Date", inplace=True)
data.head()
```

```python
data.plot()
plt.show()
```

```python
data['Adj Close'].plot()
plt.show()
```

```python
data[['High','Low']].plot()
plt.show()
```

```python
data['100ma'] = data['Adj Close'].rolling(window=100).mean()
```

```python
print(data.head())
```

```python
data['100ma'] = data['Adj Close'].rolling(window=100,min_periods=0).mean()
print(data.head())
```

```python
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1, sharex=ax1)

ax1.plot(data.index, data['Adj Close'])
ax1.plot(data.index, data['100ma'])
ax2.bar(data.index, data['Volume'])

plt.show()
```
